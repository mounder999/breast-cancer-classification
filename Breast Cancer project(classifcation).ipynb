```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Charger le dataset avec le bon dlimiteur
data = pd.read_csv('breast_cancer.csv', delimiter=';')

# Extraire les colonnes numriques correspondant aux gnes
gene_expression_data = data.iloc[:, 1:-1].apply(pd.to_numeric, errors='coerce')

### 1. Traitement des donnes manquantes
# Vrifier les valeurs manquantes
print("Valeurs manquantes avant traitement :")
print(gene_expression_data.isnull().sum())

# Remplacer les valeurs manquantes par la mdiane de chaque colonne
for column in gene_expression_data.columns:
    median = gene_expression_data[column].median()
    gene_expression_data[column].fillna(median, inplace=True)

# Vrifier aprs traitement
print("\nValeurs manquantes aprs traitement :")
print(gene_expression_data.isnull().sum())

### 2. Traitement des outliers
# Calculer Q1, Q3, et IQR pour chaque colonne
Q1 = gene_expression_data.quantile(0.25)
Q3 = gene_expression_data.quantile(0.75)
IQR = Q3 - Q1

# Calculer les bornes infrieures et suprieures
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remplacer les valeurs en dehors des bornes
for column in gene_expression_data.columns:
    gene_expression_data[column] = np.clip(gene_expression_data[column], lower_bound[column], upper_bound[column])

### 3. Standardisation des donnes
scaler = StandardScaler()
standardized_data = scaler.fit_transform(gene_expression_data)

# Convertir en DataFrame pour une manipulation plus facile
standardized_data = pd.DataFrame(standardized_data, columns=gene_expression_data.columns)

# Rsultat final : donnes nettoyes et standardises
print("\nDonnes aprs standardisation (premires lignes) :")
print(standardized_data.head())
final_data = pd.concat([data['id_sample'], standardized_data, data['pam50']],axis=1)

print(final_data.head())
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principal_component = pca.fit_transform(standardized_data)

pca_result = pd.DataFrame(data=principal_component, columns=['PC1','PC2'])
pca_result['pam50'] = final_data['pam50']

plt.figure(figsize=(10, 7)) 
for subtype in pca_result['pam50'].unique():
    subset = pca_result[pca_result['pam50'] == subtype]
    plt.scatter(subset['PC1'], subset['PC2'], label=subtype, alpha=0.7)

plt.title('PCA of Breast Cancer Gene Expression Data')
plt.xlabel('Principle Component 1 (PC1)')
plt.ylabel('Principle Component 2 (PC2)')
plt.legend(title='molecular Subtype')
plt.grid()
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

linkage_matrix = linkage(standardized_data)
plt.figure(figsize=(20, 9))
dendrogram(linkage_matrix, labels=data['id_sample'].values, p=10,
           truncate_mode='level', leaf_rotation=90, leaf_font_size=10)
plt.title('Dendrogramme of Breast Cancer Gene Expression Data')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

n_clusters = 4
model = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
cluster_labels = model.fit_predict(standardized_data)

final_data['Cluster'] = cluster_labels

# Visualize the cluster results
pca = PCA(n_components=2)
pca_result = pca.fit_transform(standardized_data)
plt.figure(figsize=(10, 7))
for cluster in range(n_clusters):
    cluster_data = pca_result[cluster_labels == cluster]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1],alpha=0.7)


plt.title('Hierarchical Agglomerative Clustering Results')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Clusters')
plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans

cluster_range = range(1, 11)
inertia = []

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=0, init='k-means++', max_iter=300)
    kmeans.fit(standardized_data)
    inertia.append(kmeans.inertia_)

# Visualize the inertia values to find the "elbow point."
plt.figure(figsize=(8, 5))
plt.plot(cluster_range, inertia, marker='o', linestyle='--', color='b')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid()
plt.show()

# Apply kmeans with optimal clusters
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=0, init='k-means++', max_iter=300)
cluster_labels = kmeans.fit_predict(standardized_data)

final_data['KMeans_clusters'] = cluster_labels
print(final_data['KMeans_clusters'])
```

```python
# Visualization of clusters after applying K-Means
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


pca = PCA(n_components=2)
pca_results = pca.fit_transform(standardized_data)

pca_df = pd.DataFrame(data=pca_results, columns=['PC1', 'PC2'])
pca_df['Cluster'] = final_data['KMeans_clusters']

plt.figure(figsize=(10, 7))
for cluster in pca_df['Cluster'].unique():
    cluster_data = pca_df[pca_df['Cluster'] == cluster]
    plt.scatter(cluster_data['PC1'], cluster_data['PC2'], label=f'Cluster {cluster}', alpha=0.7)

plt.title('K-Means Clustering Results (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Clusters')
plt.grid()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

X = data.iloc[:, 1:-1]
y = data['pam50']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# standardize the training and testing data
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)

# function to evaluate the models (KNN, Desicion Tree and Naive Bayes)
def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    return accuracy, precision, recall, f1

results = []
```

```python
# knn
for k in [3, 5, 7, 9]:
    knn_model = KNeighborsClassifier(n_neighbors=k)
    accuracy, precision, recall, f1 = evaluate_model(model=knn_model, X_train=X_train_std, X_test=X_test_std ,y_train=y_train, y_test=y_test)
    results.append({'Model': f'KNN (k={k})', 'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1-Score': f1})
```

```python
#  Desicion Tree
dt_model = DecisionTreeClassifier(random_state=42)
accuracy, precision, recall, f1 = evaluate_model(dt_model, X_train, X_test, y_train, y_test)
results.append({'Model': 'Decision Tree', 'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1-Score': f1})
```

```python
# Using Naive Bayes
nb_model = GaussianNB()
accuracy, precision, recall, f1 = evaluate_model(nb_model, X_train_std, X_test_std, y_train, y_test)
results.append({'Model': 'Naive Bayes', 'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1-Score': f1})
```

```python
# Saving results in csv file
results_df = pd.DataFrame(results)
results_df.to_csv('classification_results.csv', index=False)
print("Results saved to classification_results.csv")
```
